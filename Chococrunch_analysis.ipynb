```python
url = "https://world.openfoodfacts.org/api/v2/search"
fields = "code,product_name,brands,nutriments"
products = []

for page in range(1, 121):   # fetch up to ~12,000 records
    params = {
        "categories": "chocolates",
        "fields": fields,
        "page_size": 100,
        "page": page
    }
    r = requests.get(url, params=params)
    data = r.json().get("products", [])
    if not data: break
    for p in data:
        products.append({
            "product_code": p.get("code"),
            "product_name": p.get("product_name"),
            "brand": p.get("brands"),
            "nutriments": p.get("nutriments")
        })

df = pd.DataFrame(products)
df
```

```python
required_nutriments = [
    "energy-kcal_value",
    "energy-kj_value",
    "carbohydrates_value",
    "sugars_value",
    "fat_value",
    "saturated-fat_value",
    "proteins_value",
    "fiber_value",
    "salt_value",
    "sodium_value",
    "nova-group",
    "nutrition-score-fr",
    "fruits-vegetables-nuts-estimate-from-ingredients_100g"
]
# flatten the nested 'nutriments' dict
nutrients_df = pd.json_normalize(df['nutriments'])

# Keep only the required columns
nutrients_df = nutrients_df[required_nutriments]

# Combine with your main DataFrame
df = pd.concat([df.drop(columns=['nutriments']), nutrients_df], axis=1)

# Replace missing numeric values with 0 (or mean, if preferred)
df[required_nutriments] = df[required_nutriments].fillna(0)

print(df.isnull().sum())
```

```python
# ==========================================================
# üßπ CLEAN, NORMALIZE & INSERT DATA INTO MYSQL (FINAL SAFE VERSION)
# ==========================================================

import pandas as pd
import re
import unicodedata
from sqlalchemy import create_engine, text

# --- MySQL connection ---
engine = create_engine("mysql+pymysql://root:%40Doordie123@localhost/choco_analysis")


df["product_code"] = df["product_code"].astype(str).str.strip()

# Pad all codes to uniform length (13 digits)
df["product_code"] = df["product_code"].apply(lambda x: x.zfill(13))

# --- Normalize text ---
def normalize_text(text):
    if pd.isna(text):
        return ""
    text = unicodedata.normalize("NFKC", str(text))
    text = re.sub(r"[^\w\s'&-]", " ", text)
    text = re.sub(r"\s+", " ", text).strip().lower()
    return text

# --- Clean brand names ---
def clean_brand_name(brand):
    b = normalize_text(brand)
    if not b:
        return "Unknown"
    if re.search(r"sainsbur", b):
        return "Sainsbury's"
    if re.search(r"(m\s*&\s*s|marks\s*(and|&)\s*spencer)", b):
        return "M&S"
    if re.search(r"nestl", b):
        return "Nestle"
    if re.search(r"aldi", b):
        return "Aldi"
    if re.search(r"lidl", b):
        return "Lidl"
    if re.search(r"tesco", b):
        return "Tesco"
    b = re.sub(r"^(by\s+)", "", b)
    b = re.split(r"[,;/]", b)[0].strip()
    return b.title()

# --- Clean columns ---
df['product_name'] = df.get('product_name', pd.Series(['Unnamed Product'] * len(df)))
df['brand'] = df.get('brand', pd.Series(['Unknown'] * len(df))).astype(str).apply(clean_brand_name)
df['brand'] = df['brand'].fillna('Unknown Brand')

# --- Standardize product codes ---
max_len = df['product_code'].astype(str).str.len().max()
df['product_code'] = df['product_code'].astype(str).str.zfill(max_len)

# --- Helper: keep only existing columns ---
def select_existing(df, cols):
    return [c for c in cols if c in df.columns]

# --- Split DataFrames safely ---
df_product = df[select_existing(df, ['product_code', 'product_name', 'brand'])].drop_duplicates()

df_nutrient = df[select_existing(df, [
    'product_code',
    'energy_kcal_value', 'energy_kj_value',
    'carbohydrates_value', 'sugars_value',
    'fat_value', 'saturated_fat_value',
    'proteins_value', 'fiber_value',
    'salt_value', 'sodium_value',
    'fruits_vegetables_nuts_estimate_from_ingredients_100g',
    'nutrition_score_fr', 'nova_group'
])].copy()

df_derived = df[select_existing(df, [
    'product_code',
    'sugar_to_carb_ratio', 'calorie_category',
    'sugar_category', 'is_ultra_processed'
])].copy()

# --- Safe MySQL UPSERT ---
def upsert_table(df, table_name, unique_key='product_code'):
    """Safely UPSERT or INSERT IGNORE if only key column exists."""
    if df.empty:
        print(f"‚ö†Ô∏è Skipping {table_name}: No data.")
        return

    with engine.begin() as conn:
        for _, row in df.iterrows():
            cols = ', '.join(df.columns)
            vals = ', '.join([f":{c}" for c in df.columns])

            # ‚úÖ Case 1: multiple columns ‚Üí UPSERT
            if len(df.columns) > 1:
                updates = ', '.join([f"{c}=VALUES({c})" for c in df.columns if c != unique_key])
                query = text(f"""
                    INSERT INTO {table_name} ({cols})
                    VALUES ({vals})
                    ON DUPLICATE KEY UPDATE {updates};
                """)
            # ‚úÖ Case 2: only one column ‚Üí INSERT IGNORE
            else:
                query = text(f"""
                    INSERT IGNORE INTO {table_name} ({cols})
                    VALUES ({vals});
                """)

            conn.execute(query, dict(row))

    print(f"‚úÖ Data inserted/updated for {table_name} ({len(df)} rows).")

# --- Execute ---
upsert_table(df_product, "product_info")
upsert_table(df_nutrient, "nutrient_info")
upsert_table(df_derived, "derived_metrics")

print("\nüéâ All data cleaned, standardized, and uploaded successfully!")

```

```python
queries_product = {
    "Count products per brand": """
        SELECT brand, COUNT(*) AS product_count
        FROM product_info
        GROUP BY brand
        ORDER BY product_count DESC;
    """,
    "Count unique products per brand": """
        SELECT brand, COUNT(DISTINCT product_name) AS unique_product_count
        FROM product_info
        GROUP BY brand
        ORDER BY unique_product_count DESC;
    """,
    "Top 5 brands by product count": """
        SELECT brand, COUNT(*) AS product_count
        FROM product_info
        GROUP BY brand
        ORDER BY product_count DESC
        LIMIT 5;
    """,
    "Products with missing product name": """
        SELECT *
        FROM product_info
        WHERE product_name IS NULL
           OR product_name = ''
           OR product_name = 'Unnamed Product';
    """,
    "Number of unique brands": """
        SELECT COUNT(DISTINCT brand) AS unique_brands
        FROM product_info;
    """,
    "Products with code starting with 3": """
        SELECT *
        FROM product_info
        WHERE CAST(product_code AS CHAR) LIKE '3%%';
    """
}

# Safe execution loop
for name, query in queries_product.items():
    print(f"\nüîπ {name}")
    try:
        df_result = pd.read_sql(query, con=engine)
        if df_result.empty:
            print("‚ö†Ô∏è No results found.")
        else:
            display(df_result.head())
    except Exception as e:
        print(f"‚ùå Error in '{name}': {e}")
```

```python
queries_nutrient = {
    # 1 Top 10 products with highest kcal
    "Top 10 products with highest energy_kcal_value": """
    SELECT 
    ni.product_code,
    MAX(ni.energy_kcal_value) AS energy_kcal_value
    FROM nutrient_info ni
    JOIN product_info pi 
        ON ni.product_code = pi.product_code
    GROUP BY ni.product_code, pi.product_name, pi.brand
    ORDER BY energy_kcal_value DESC
    LIMIT 10;
    """,

    # 2 Average sugars per nova group
    "Average sugars_value per nova_group": """
        SELECT 
        COALESCE(nova_group, 'Unknown') AS nova_group,
        ROUND(AVG(sugars_value), 2) AS avg_sugars
        FROM nutrient_info
        GROUP BY COALESCE(nova_group, 'Unknown')
        ORDER BY nova_group;

    """,

    # 3 Count of products with fat > 20g
    "Count products with fat_value > 20g": """
        SELECT COUNT(*) AS high_fat_products
        FROM nutrient_info
        WHERE fat_value > 20;
    """,

    # 4 Average carbohydrates per product
    "Average carbohydrates_value per product": """
        SELECT 
        product_code,
        ROUND(AVG(carbohydrates_value), 2) AS avg_carbohydrates
        FROM nutrient_info
        WHERE carbohydrates_value IS NOT NULL
        GROUP BY product_code
        ORDER BY avg_carbohydrates DESC;
    """,

    # 5 Products with sodium > 1g
    "Products with sodium_value > 1g": """
        SELECT DISTINCT product_code, sodium_value
        FROM nutrient_info
        WHERE sodium_value > 1
        ORDER BY sodium_value DESC;

    """,

    # 6 Count products with non-zero fruits/vegetables/nuts
    "Count products with non-zero fruits_vegetables_nuts content": """
        SELECT COUNT(*) AS non_zero_fvn_products
        FROM nutrient_info
        WHERE fruits_vegetables_nuts_estimate_from_ingredients_100g > 0;
    """,

    # 7 Products with kcal > 500
    "Products with energy_kcal_value > 500": """
        SELECT DISTINCT product_code, energy_kcal_value
        FROM nutrient_info
        WHERE energy_kcal_value > 500
        ORDER BY energy_kcal_value DESC;
    """
}

# --- ‚úÖ Execute and Display ---
for name, query in queries_nutrient.items():
    print(f"\nüîπ {name}")
    try:
        df_result = pd.read_sql(query, con=engine)
        if df_result.empty:
            print("‚ö†Ô∏è No results found.")
        else:
            display(df_result.head(10))  # show top 10 rows for context
    except Exception as e:
        print(f"‚ùå Error in '{name}': {e}")
```

```python
queries_derived = {
    # 1 Count products per calorie category
    "Count products per calorie_category": """
        SELECT calorie_category, COUNT(*) AS product_count
        FROM derived_metrics
        WHERE calorie_category IS NOT NULL
        GROUP BY calorie_category
        ORDER BY product_count DESC;
    """,

    # 2 Count of High Sugar products
    "Count of High Sugar products": """
        SELECT COUNT(*) AS high_sugar_products
        FROM derived_metrics
        WHERE sugar_category = 'High Sugar';
    """,

    # 3 Average sugar_to_carb_ratio for High Calorie products
    "Average sugar_to_carb_ratio for High Calorie products": """
        SELECT ROUND(AVG(sugar_to_carb_ratio), 3) AS avg_sugar_to_carb_ratio
        FROM derived_metrics
        WHERE calorie_category = 'High'
          AND sugar_to_carb_ratio IS NOT NULL;
    """,

    # 4 Products that are both High Calorie and High Sugar
    "Products that are both High Calorie and High Sugar": """
        SELECT product_code, sugar_to_carb_ratio
        FROM derived_metrics
        WHERE calorie_category = 'High'
          AND sugar_category = 'High Sugar';
    """,

    # 5 Number of products marked as ultra-processed
    "Number of products marked as ultra-processed": """
        SELECT COUNT(*) AS ultra_processed_products
        FROM derived_metrics
        WHERE is_ultra_processed = 'Yes';
    """,

    # 6 Products with sugar_to_carb_ratio > 0.7
    "Products with sugar_to_carb_ratio > 0.7": """
        SELECT DISTINCT product_code, sugar_to_carb_ratio, calorie_category, sugar_category
        FROM derived_metrics
        WHERE sugar_to_carb_ratio > 0.7
        ORDER BY sugar_to_carb_ratio DESC;
    """,

    # 7 Average sugar_to_carb_ratio per calorie category
    "Average sugar_to_carb_ratio per calorie_category": """
        SELECT calorie_category,
               ROUND(AVG(sugar_to_carb_ratio), 3) AS avg_sugar_to_carb_ratio
        FROM derived_metrics
        WHERE sugar_to_carb_ratio IS NOT NULL
        GROUP BY calorie_category
        ORDER BY avg_sugar_to_carb_ratio DESC;
    """
}


# --- ‚úÖ Execute and Display ---
for name, query in queries_derived.items():
    print(f"\nüîπ {name}")
    try:
        df_result = pd.read_sql(query, con=engine)
        if df_result.empty:
            print("‚ö†Ô∏è No results found.")
        else:
            display(df_result.head(10))  # Display top 10 for readability
    except Exception as e:
        print(f"‚ùå Error in '{name}': {e}")
```

```python
queries_join = {
    # 1 Top 5 brands with most High Calorie products
    "Top 5 brands with most High Calorie products": """
        SELECT pi.brand,
               COUNT(*) AS high_calorie_products
        FROM product_info pi
        JOIN derived_metrics dm ON pi.product_code = dm.product_code
        WHERE dm.calorie_category = 'High'
        GROUP BY pi.brand
        ORDER BY high_calorie_products DESC
        LIMIT 5;
    """,

    # 2 Average energy-kcal_value for each calorie_category
    "Average energy-kcal_value for each calorie_category": """
        SELECT dm.calorie_category,
               ROUND(AVG(ni.energy_kcal_value), 2) AS avg_energy_kcal
        FROM derived_metrics dm
        JOIN nutrient_info ni ON dm.product_code = ni.product_code
        WHERE ni.energy_kcal_value IS NOT NULL
        GROUP BY dm.calorie_category
        ORDER BY avg_energy_kcal DESC;
    """,

    # 3 Count of ultra-processed products per brand
    "Count of ultra-processed products per brand": """
        SELECT pi.brand,
               COUNT(*) AS ultra_processed_count
        FROM product_info pi
        JOIN derived_metrics dm ON pi.product_code = dm.product_code
        WHERE dm.is_ultra_processed = "Yes"
        GROUP BY pi.brand
        ORDER BY ultra_processed_count DESC;
    """,

    # 4 Products with High Sugar and High Calorie along with brand
    "Products with High Sugar and High Calorie along with brand": """
        SELECT DISTINCT pi.product_name,
               pi.brand,
               dm.sugar_to_carb_ratio,
               dm.calorie_category,
               dm.sugar_category
        FROM product_info pi
        JOIN derived_metrics dm ON pi.product_code = dm.product_code
        WHERE dm.sugar_category = 'High Sugar'
          AND dm.calorie_category = 'High'
        ORDER BY dm.sugar_to_carb_ratio DESC;
    """,

    # 5 Average sugar content per brand for ultra-processed products
    "Average sugar content per brand for ultra-processed products": """
        SELECT pi.brand,
               ROUND(AVG(ni.sugars_value), 2) AS avg_sugar
        FROM product_info pi
        JOIN nutrient_info ni ON pi.product_code = ni.product_code
        JOIN derived_metrics dm ON pi.product_code = dm.product_code
        WHERE dm.is_ultra_processed = "Yes"
          AND ni.sugars_value IS NOT NULL
        GROUP BY pi.brand
        ORDER BY avg_sugar DESC;
    """,

    # 6 Number of products with fruits/vegetables/nuts content in each calorie_category
    "Number of products with fruits/vegetables/nuts content in each calorie_category": """
        SELECT dm.calorie_category,
               COUNT(*) AS fruit_veg_nut_products
        FROM derived_metrics dm
        JOIN nutrient_info ni ON dm.product_code = ni.product_code
        WHERE ni.fruits_vegetables_nuts_estimate_from_ingredients_100g > 0
        GROUP BY dm.calorie_category
        ORDER BY fruit_veg_nut_products DESC;
    """,

    # 7 Top 5 products by sugar_to_carb_ratio with their calorie and sugar category
    "Top 5 products by sugar_to_carb_ratio with their calorie and sugar category": """
        SELECT DISTINCT pi.product_name,
               pi.brand,
               dm.sugar_to_carb_ratio,
               dm.calorie_category,
               dm.sugar_category
        FROM product_info pi
        JOIN derived_metrics dm ON pi.product_code = dm.product_code
        WHERE dm.sugar_to_carb_ratio IS NOT NULL
        ORDER BY dm.sugar_to_carb_ratio DESC
        LIMIT 5;
    """
}


# ---  Execute & Display Results ---

for name, query in queries_join.items():
    print(f"\nüîπ {name}")
    try:
        df_result = pd.read_sql(query, con=engine)
        if df_result.empty:
            print("‚ö†Ô∏è No results found.")
        else:
            display(df_result.head(10))
    except Exception as e:
        print(f"‚ùå Error in '{name}': {e}")
```

```python
brand_nutrition = (
    df.groupby('brand', dropna=True)
      .agg(
          avg_kcal=('energy_kcal_value', 'mean'),
          avg_sugar=('sugars_value', 'mean'),
          product_count=('product_code', 'nunique')
      )
      .reset_index()
)

# Filter out tiny brands with too few products so results are meaningful
brand_nutrition = brand_nutrition[brand_nutrition['product_count'] >= 5]

# Top 10 by avg sugar
top_sugar = brand_nutrition.sort_values('avg_sugar', ascending=False).head(10)

plt.figure(figsize=(10,4))
sns.barplot(
    data=top_sugar,
    x='brand',
    y='avg_sugar'
)
plt.title("Top 10 Brands by Avg Sugar (g/100g)")
plt.xlabel("Brand")
plt.ylabel("Avg Sugar (g)")
plt.xticks(rotation=45, ha='right')
plt.show()

# Top 10 by avg kcal
top_kcal = brand_nutrition.sort_values('avg_kcal', ascending=False).head(10)

plt.figure(figsize=(10,4))
sns.barplot(
    data=top_kcal,
    x='brand',
    y='avg_kcal'
)
plt.title("Top 10 Brands by Avg Calories (kcal/100g)")
plt.xlabel("Brand")
plt.ylabel("Avg kcal/100g")
plt.xticks(rotation=45, ha='right')
plt.show()
```
